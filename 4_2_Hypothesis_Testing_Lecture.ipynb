#Hypothesis Testing

In today's lecture, we will be looking at how we are able to test claims using data via Hypothesis Testing. 

**A Hypothesis** is a claim that we want to test. *Ex. There's a pot of gold at the end of a rainbow* Using data, we try to make a claim and using statistical techniques to inform those claims. 

**A Null Hypothesis** is a default claim that we assume is true. *Ex. The average birth weight of humans is 5.5 lbs and 10 lbs* When we want to challenge the null hypothesis, we need to come up with a claim that we think is true instead. This is known as an *Alternative Hypothesis*.

**An Alternative Hypothesis** is simply the claim we test out to see if it's true.

In the passage below, find out what the null and alternative hypotheses are: 

It is believed that a candy machine makes chocolate bars that are on average 5g. A worker claims that the machine after maintenance no longer makes 5g bars.

##Testing
The outcomes of a test are either Reject Null or fail to reject Null. In statistics, we don't say that we have proven something after one hypothesis test as proving a claim is alot more rigorous. 

To actually do the testing, we rely on a **Test Statistic**.

**A Test Statistic** is a quantity that we calculate from our sample data. In the example above, if we were testing on 50 chocolate bars, our test statistic would be the average weight of these groups of chocolate bars. Once we calculate our test statistic, we determine whether or not this statistic is **statistically significant**. Statistical significance is essentially where we draw the line to see if our test statistic is enough to reject a null hypothesis.

Going back to the example of the chocolate bars, if we were to get groups of 50 chocolate bars and measure their averages to be 5.12 g, 5.72 g & 7.23 g, upon measuring 7.23 g as the average weight of the chocolate bar, we would have evidence to reject the null of the average weight of the chocolate bars being 5 g.

The degree to which we are sure of rejecting our null hypothesis is the **Level of Confidence**. Usually we place the level of confidence at a high number (95% or 99%) to say that our test statistic is a believable grounds to reject our null.

#Z Test

The Z test assumes the distribution of the data follows a normal distribution, assumes the population to be tested is big (>30) and the population standard deviation and mean is known. 

We calculate a Z statistic as: 

$$Z = \frac{x-\mu}{\frac{\sigma}{\sqrt n}}$$

Where $\mu$ and $\sigma$ are the known population mean and standard deviation and x is our test statistic (Usually an average).

Ex. It is widely believed that the average diameter of straws is 3 mm with a standard deviation of 1 mm. We measure the diameter of several straws and get the diameters as follows: [2.5, 2, 3, 3.3, 4.6, 2.3, 7.2, 4.5, 1.2, 3.4, 1.1, 2.9, 3.6, 3.9, 4.2, 4.5, 4.7, 4.8, 2.9, 3.3, 4.1, 2.3, 2.7, 2.9, 3.3, 3.2, 4.6, 4.8, 3.3, 4.6, 4.6, 5.7, 5.7, 5.8, 6.0, 2.3, 1.0, 7.1, 6.9] perform  a z test with 95% confidence to see if we have evidence to reject this claim:



```
import numpy as np

diameters =  [2.5, 2, 3, 3.3, 4.6, 2.3, 7.2, 4.5, 1.2, 3.4, 1.1, 2.9, 3.6, 3.9, 4.2, 4.5, 4.7, 4.8, 2.9, 3.3, 4.1, 2.3, 2.7, 2.9,3.3, 3.2, 4.6, 4.8, 3.3, 4.6, 4.6, 5.7, 5.7, 5.8, 6.0, 2.3, 1.0, 7.1, 6.9]
n = len(diameters)
diameters = np.array(diameters)
x = np.mean(diameters)
```

```
def z_score(x, mu, sd, n):
  z = (x-mu)/(sd/n**0.5)
  if abs(z)>1.96:
    print(f"Our value waz {abs(z)} which is greater than 1.96 and we evidence to reject the null hypothesis at 95% confidence")
  else:
    print(f"Our z value was {abs(z)} which is less than 1.96 and we fail to reject the null")

z_score(x, 3, 1, n)
```

If we have 2 samples with a known population standard deviation and mean, we use a 2 sample z test for means to see if we have a difference between groups.

The test statistic for this is calculated as:

$$Z = \frac{(x_1 - x_2) - (\mu_1 - \mu_2)}{\sqrt{\frac{\sigma_1^2}{n_1} + \frac{\sigma_2^2}{n_2}}}$$

```
pop1 = np.random.normal(1,5,40)
diameters =  [2.5, 2, 3, 3.3, 4.6, 2.3, 7.2, 4.5, 1.2, 3.4, 1.1, 2.9, 3.6, 3.9, 4.2, 4.5, 4.7, 4.8, 2.9, 3.3, 4.1, 2.3, 2.7, 2.9,3.3, 3.2, 4.6, 4.8, 3.3, 4.6, 4.6, 5.7, 5.7, 5.8, 6.0, 2.3, 1.0, 7.1, 6.9]
x1 = np.mean(pop1)
x2 = np.mean(np.array(diameters))
mu1 = 1
mu2 = 3
sd1 = 5
sd2 = 1
n1 = 40
n2 = len(diameters)

def two_sample_z(x1, x2, mu1, mu2, sd1, sd2, n1, n2):
  z = ((x1-x2) - (mu1 - mu2)) / ((sd1**2/n1)+(sd2**2/n2))**0.5
  if abs(z)>1.96:
    print(f"Our value waz {abs(z)} which is greater than 1.96 and we evidence to reject the null hypothesis at 95% confidence")
  else:
    print(f"Our z value was {abs(z)} which is less than 1.96 and we fail to reject the null")

two_sample_z(x1, x2, mu1, mu2, sd1, sd2, n1, n2)
```

#t - Test

This is the more widely used test as it is more accurate than a z test. The 1 sample t-test also assumes that the data is normally distributed, each measurement is independent of one another and it must not contain any outliers. The t test statistic for 1 sample is:

$$t = \frac{\bar{x}-\mu}{\frac{s}{\sqrt n}}$$

where $\bar{x}$ is the sample mean, $\mu$ is the population mean, s is the sample variance and n is the sample size. 

ex

```
import pandas as pd
import scipy.stats as sc
url = "https://condor.depaul.edu/sjost/it223/datasets/hyp-exam-scores.xls"
df = pd.read_excel(url)
df.head()
```

```
score = np.array(df['ExamScores'])
sc.ttest_1samp(score, 60)
```

When we have 2 samples, depending on the sample charactersitic of the sample we will use a corresponding t test for it. In the case where we have 2 samples with equal variance (pooled t test), our test statistic is:

$$t = \frac{\bar{x_1} - \bar{x_2}}{\sqrt{s^2(\frac{1}{n_1} + \frac{1}{n_2})}}$$

where:

$$s = \frac{\sum_{i=1}^{n_1} (x_i - \bar{x_1})^2 + \sum_{j=1}^{n_2} (x_j - \bar{x_2})^2}{n_1 + n_2 -2} $$

In cases where we have don't have equal variance, we use the Welch-Satterthwaite test:

$$t= \frac{\bar{x_1} - \bar{x_2}}{\sqrt{\frac{s_1^2}{n_1} + \frac{s_2^2}{n_2}}}$$

```
url = "http://www.utsc.utoronto.ca/~butler/c32/drp.txt"
df = pd.read_table(url, delimiter= " ")
df
```

```
df1 = df.groupby('group')
df1.boxplot()
```

#Chi Square Test

The tests above have mainly worked with numerical data. In the case where we have categorical data and want to measure a frequency of categorical data to see if it's statistically significant, we use a chi square test.

$$X_c^2 = \sum \frac{(O_i - E_i)^2}{E_i}$$

Where $O_i$ is the observed frequency of our data and $E_i$ is the expected frequency.

The Chi Square Statistic, follows the chi square distribution where your degrees of freedom (df) is n-1.


image.png



Once we calculate our test statistic, we use our degrees of freedom and level of confidence to look up the critical value in the chi square table. If our statistic is higher than the value listed in the table, then our value is statistically significant.

```
values = [72, 44, 34, 53]
Expected = 50

chi = 0
for value in values:
  x = (value-Expected)**2/Expected
  chi = chi + x
print(chi) 
```

#ANOVA

Finally, the last test we will talk about is the Analysis Of Variance or ANOVA. With ANOVA we have the One Way ANOVA test and the Two Way ANOVA test. 

The One Way ANOVA tests to see if a single categorical variable has significance on a dependent continuous variable.

The Two Way ANOVA tests to see if two or more categorical variables have significance on a single continuous variable.

```
url = "https://data.montreal.ca/dataset/1e153f9f-3930-4133-8d35-ade2e8f7e7e3/resource/ef2a8162-0644-47e7-bd03-bea33f14a5d2/download/projetpietonnisation2017.csv"
df = pd.read_csv(url)
df.head()
```

```
df1 = df[["VOIE_CYCLABLE", "LONGUEUR_TRONCON"]]
df1.head()
```

```
Yes = np.array(df1[df1["VOIE_CYCLABLE"]=="Oui"]["LONGUEUR_TRONCON"])
No = np.array(df1[df1["VOIE_CYCLABLE"]=="Non"]["LONGUEUR_TRONCON"])

sc.f_oneway(Yes, No)
```

```
import statsmodels.api as sm
from statsmodels.formula.api import ols
```

```
df2 = df[["VOIE_CYCLABLE", "PASSAGE_BUS", "LONGUEUR_TRONCON"]]
#Perform Two Way ANOVA
model = ols('LONGUEUR_TRONCON ~ C(VOIE_CYCLABLE) + C(PASSAGE_BUS) + C(VOIE_CYCLABLE):C(PASSAGE_BUS)', data = df2).fit()
sm.stats.anova_lm(model, typ=2)
```
