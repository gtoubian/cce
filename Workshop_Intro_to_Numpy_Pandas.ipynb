#Logistic Regression

In this exercise, we will create a logistic regression model that will try to classify whether a species is setosa or not given sepal length. In logistic regression, we essentially get our model to assign a probability (between 0 and 1) to a given input and based off the probability calculated, the model will output its classfication. Here is a picture representing what a logistic regression model is essentially doing.

image.png


We try to fit our model to the sigmoid curve with:

$$\frac{1}{1+e^{-[w_0+w_1x]}}$$

In the cell below, Write a python function for the sigmoid function.


````
def sigmoid(w,x):
  return 1/(1 + np.exp(-1*(w[0]+w[1]*x)))
````

Next we have to include our loss function. Before with Linear Regression, we used Ordinary Least Squares and reduced the Mean Squared Error (MSE) of our model to a minimum in order to find values for m and b. With this classification algorithm, we use a differnt function called the **Binary Cross Entropy** in order to help us select values of w for which our cross entropy achieves a minimum. Keeping in mind that this model will tell us whether a flower will be a setosa or not, our output variables in Y will be either 1 (Setosa) or a 0 (Not a Setosa). 

In the cell below, create a numpy array, X, with all the sepal lengths and create another array, Y, where the first 50 entries are 1s and the next 100 entries as 0.

Run the Cross Entropy function below:

Now, that we have our cost function defined, we use a technique called **gradient descent** where we update and find values of w, in order to get the derivative of our cost function and minimize our error. Run all cells below in order to train the logistic regression model off our data.

````
def grad(w, X, Y):
    y_pred = sigmoid(w,X)
    g = [0]*3
    g[0] = -1 * sum(Y*(1-y_pred) - (1-Y)*y_pred)
    g[1] = -1 * sum(Y*(1-y_pred)*X - (1-Y)*y_pred*X)
    return g
````

````
def descent(w_new, w_prev, lr):
    print(w_prev)
    print(cost(w_prev, X, Y))
    j=0
    while True:
        w_prev = w_new
        w0 = w_prev[0] - lr*grad(w_prev, X, Y)[0]
        w1 = w_prev[1] - lr*grad(w_prev, X, Y)[1]
        w_new = [w0, w1]
        print(w_new)
        print(cost(w_new, X, Y))
        if (w_new[0]-w_prev[0])**2 + (w_new[1]-w_prev[1])**2 <pow(10,-20):
            return w_new
        
        if j>1000: 
            return w_new
        j+=1
````

##Reverse Words in a String

Given a string, you need to reverse the order of characters in each word within a sentence while still preserving whitespace and initial word order.

```
Input: "Let's take LeetCode contest"
Output: "s'teL ekat edoCteeL tsetnoc"
```




````
def backwards(string):
  words = string.split(" ")
  newwords = [i[::-1] for i in words]
  newstring = " ".join(newwords)
  return newstring
string = "Let's take LeetCode contest"
backwards(string)
````

##Length of Last Word
Given a string s consists of some words separated by spaces, return the length of the last word in the string. If the last word does not exist, return 0.

A word is a maximal substring consisting of non-space characters only.



```
Input: s = "Hello World"
Output: 5
```



````
def lastword(x): 
  length = 0
  for n in range(len(x)):
    if x[n] == " ":
      length = 0
    else:
      length += 1
  return length

lastword('Hello World')
````
