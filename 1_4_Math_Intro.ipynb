# Calculus

To put it in very simple terms, calculus is the mathematics that describes infinitesimal change. If you drive, you'll see an example of calculus in your car's very own speedometer. **A measure of your speed as a change in distance over a change in time**

In calculus, you will see two major techniques - Differentiation and Integration. Differential Calculus essentially tells us how an infinitesimal change in a variable changes an output. Integral Calculus gives us a measure of the total change of our output given an infinitesimal change in a variable. Let's dive into each further for you to get a better understanding of what I have mentioned.

### Differential Calculus 

Differential Calculus is the mathematics behind how a change in one variable changes an output. Going back to the example of the car speedometer, speed is a measure of a change in distance over a change in time:

<center>V = $\frac{ds}{dt}$<center>

<l>**NOTE:** The d symbol above stands for "delta" this is a term used often in math and physics to describe change.<l>


Another way we can say this, is that speed is the **derivative** of distance with respect to time. Derivatives are a powerful tool that not only tell you how a function changes, but can also tell you the value of a variable for which a function reaches a maximum or minimum value as the derivative will be equal to 0. The slope of a line is another example of a derivative as it is the change in y over the change in x.

Formally, the equation for the derivative of a Function is written as:

$$f'(x) = \lim_{h \to 0} \frac{f(x + h) - f(x)}{h}$$

Let's do an example to visualize finding a maximum/minimum of a function so that you can better understand.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
##Don't worry about the package imports for now.

def f(x):
  return 9 - (x-3)**2
## Here, I defined a function in python where f(x) = 9 - (x-3)^2

def fprime(x):
  h = 1e-5
  return (f(x+h) - f(x))/h
## Here, I defined the derivative function as f' or fprime.

x = np.linspace(-3,9,19)

fig, axs = plt.subplots(2)
plt.figure(figsize=(30,20))
fig.suptitle('Plots of the function f(x) and its derivative, fprime(x)')
axs[0].plot(x, f(x))
axs[1].plot(x, fprime(x))
```

Here, you can see that at the value of x = 3, our function f(x), reaches its maximum value of 9 whereas our derivative hits a value of 0 at x = 3. We will be using derivatives a good amount when we get into talking about reducing error in our models.

###Integral Calculus
Integral Calculus is the opposite of differentiation. Rather than the slope of a curve, the integral is the area underneath the curve. You can think of an integral as a measure of accumulated change - the end result of a large number of changes to some quantity of interest that have built up over time (or along some dimension more generally) and whether those changes are positive or negative.

Going back to the Speedometer example, if we were to integrate the function for speed over some interval of time, we would get the **Total** distance the car covered in that amount of time.

While there are many ways to calculate an integral, I will use a Riemann Sum to represent an integral.

 $$\int_a^b f(x) dx \approx \lim_{n \to \infty} \sum_{i=1}^n f(x_i)*\Delta x $$  

 $$\Delta x = \frac{b-a}{n} $$

 $$x_i = a + i \Delta x$$

 Since there is not much of a clear cut way to get the area under a curve, what this approximation does is fit a "n" number of rectangles under a curve and gets the sum of the individual rectangles. The more rectangles, the more accurate the integral. Here is a picture visually showing how the integral becomes more accurate as the number of rectangles goes up.

#Linear Algebra

Linear algebra is a branch of math that deals with solving linear equations in the form of:

$$a_1x_1 + a_2x_2+a_3x_3+...+a_nx_n = b$$

by using matrices and vectors. 

Let's do an example. Say you wanted to solve the following system of equations for values of a,b and c.

2a + b + 2c = 20

a + 5b + 3c = 38

4a + b + 2c = 26

Sure, we could solve it by substitution but that would take far too long. Instead we can represent this system of linear equations as a matrix multiplied by a vector as:

$$\begin{bmatrix} 2 & 1 & 2 \\ 1 & 5 & 3 \\ 4 & 1 & 2 \end{bmatrix} \begin{bmatrix} a \\ b \\ c \end{bmatrix} = \begin {bmatrix} 20\\ 38\\ 26 \end{bmatrix} $$

If you're unsure of how the above equation is equal to the system of equations we had before, this is an example of matrix multiplication. This video by PatrickJMT should cover it visually.

https://www.youtube.com/watch?v=YtMYfvypgM4&ab_channel=patrickJMT

But essentially, you multiply each row of the left matrix by the columns of the matrix on the right. The matrix on the right with only 1 column can also be referred to as a vector. Try doing the matrix multiplication by hand after watching the video.

To solve the above equation, there are a number of procedure. The most basic of them is the **Gauss Jordan Elimination** where we convert the equation to the form of: 

$$\begin{bmatrix} 1 & 0 & 0 \\ 0 & 1 & 0 \\ 0 & 0 & 1 \end{bmatrix} \begin{bmatrix} a \\ b \\ c \end{bmatrix} = \begin {bmatrix} A\\ B\\ C \end{bmatrix} $$

And in this case, we have a = A, b = B and c = C. 

Here is a video on how actually solving the system of equations looks like by hand.  
https://www.youtube.com/watch?v=eYSASx8_nyg&ab_channel=TheOrganicChemistryTutor

Luckily since we have python, we won't have to solve these by hand and instead get the computer to solve it as in the example below. 




```python
##We'll use numpy to solve the matrix version of our system of equations

Matrix = np.array([[2,1,2], [1,5,3], [4,1,2]])
Solutions = np.array([20, 38, 26])
x = np.linalg.solve(Matrix, Solutions)
x
```

After running the above code to solve the system of equations we obtain in order of a = 3, b = 4 & c = 5 as expected.

As a last point on matrices, some matrices also have an interesting property where they can be rewritten in the form of:

$$\begin{bmatrix} x & 0 & 0 \\ 0 & y & 0 \\ 0 & 0 & z \end{bmatrix}$$

where x, y and z are non zero entries and can also be called **Eigenvalues** of the matrix.

##Statistics

The last main area we will cover is Mathematical Statistics. Statistics uses a wide variety of math in its application. You may have already seen the most basic calculations involved in statistics such as Mean, Median, Mode, Variance,Standard Deviation and Interquartile Range. Let's quickly cover these.

**The Mean** of a set of numbers is also referred to as the average value. In Statistics, it can also be referred to as an **Expectation Value**.

The formula for the mean can be written as:

$$\mu = \frac {{\sum_{i=1}^n}x_i}{n}$$

where, x are a set of numbers and n is the total amount of numbers in the set.

Or as:

$$\mu = \sum_{i=1}^n p_ix_i$$

where p is the probability of getting a result x. We will dive more into this in our probability module.

Ex. Given a set of numbers, (1, 0, 4, 5, 6, 8, 10, 6) we can calculate the average as:

$$\mu = \frac {1+0+4+5+6+8+10+6}{8} = 5$$\

**The Median** of an ***ordered*** set of numbers refers to the number that is in the middle of the set. (NOTE: Ordered set just means that the numbers are in ascending order, starting from the lowest number and moving towards the greatest) 

For Example to find the median of this set, (1,4,3,5,8), we must first order it as: (1,3,4,5,8) and we find that the median is 4. Another way we can represent the median is by saying that it is the number at which 50% of the total numbers in the set are equal to or less than it.

**The Variance and Standard Deviation** are a measure of the spread of the set of numbers from the mean. The variance  can be calculated as:

$$var = \sigma^2 = \frac {\sum_{i=1}^n (x_i - \mu)^2}{n}$$

The Standard Deviation is the square root of the variance. Note: The mean and standard deviation are very susceptible to outliers in your data.

**The Interquartile Range** is a measure of the spread about the median and is more resistant to being affected by outliers. Going back to what was said about the median being the number at which 50% of the total numbers are equal to or less than it, we can also refer to the median as the ***Second Quartile***. The interquartile range is obtained by subtracting the third quartile from the first quartile. The first quartile is the number at which 25% of the total numbers in a set are equal to or less than it while the third quartile is the number at which this is true for 75% of the numbers.

Example: Given an ordered set of data, (2,4,4,5,6,7,8) the number 4 is the first quartile, 5 is the second quartile and 7 is the third quartile. Here, is a picture to represent this visually:
