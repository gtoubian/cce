## Seaborn Zoo
Since the project and last weeks remaining notebooks are plenty to have on your plate we'll just be taking a brief tour on seaborns main options and how you can use them to approach a basic problem. Having them in your backpocket may be useful in your future data science career.

```
import seaborn as sns
sns.get_dataset_names()
```

```
penguins = sns.load_dataset('penguins')
penguins
```

```
penguins = penguins.dropna()
```

The types of graphs you want to use depends on what data you have and which you want to visualize
For simple categorical data a bar chart can do the trick


```
#Categorical 
sns.barplot(x='island', y='sex', data=penguins)
```

```
#Categorical 
sns.barplot(x='island', y='flipper_length_mm', data=penguins)
```

```
#Multiple Categorical
sns.catplot(data=penguins, kind="bar", x="island", y='flipper_length_mm', hue="sex")
```

```
sns.catplot(data=penguins, x="island", y='flipper_length_mm', hue="sex")
```

Say you have quantitative now (numerical values)

```
#Quantitative
sns.histplot(data=penguins, x="flipper_length_mm")
```

```
#Multi Quantitative
sns.scatterplot(data=penguins, x="flipper_length_mm", y="bill_length_mm")
```

```
#Multi Quantitative w/ 1 Categorical
sns.scatterplot(data=penguins, x="flipper_length_mm", y="bill_length_mm", hue = 'sex')
```

```
# 1 Categorical and 1 Quantitative 
sns.boxplot(x="island", y="flipper_length_mm",
            hue="sex",
            data=penguins)
```

```
flights = sns.load_dataset("flights")
flights = flights.pivot("month", "year", "passengers")
ax = sns.heatmap(flights)
```
