# 1992-1993 Chicago Bulls Player Stats
For this project, I will scrape the data from the REALGM Basketball website, specifically relating to the statistics (averages) for each individual player on the 1992-1993 Chicago Bulls.

**Step 1: Import the packages**

```python
import requests
from bs4 import BeautifulSoup as bs
```

**Step 2: Make the request**

```python
r = requests.get("https://basketball.realgm.com/nba/teams/Chicago-Bulls/4/Stats/1993/Averages/Qualified/points/All/desc/1/Regular_Season")
soup = bs(r.content, 'html.parser')
content = soup.prettify()
content
```

**Step 3: Scrape the data (in table form) and create a Pandas DataFrame**

```python
import pandas as pd
table = soup.find('table', {'class': 'tablesaw', 'data-tablesaw-mode': 'swipe'})
columns = table.find("thead").find_all("th")
column_names = [c.string for c in columns]

table_rows = table.find("tbody").find_all("tr")
l = []
for tr in table_rows:
  td = tr.find_all('td')
  row = [str(tr.get_text()).strip() for tr in td]
  l.append(row)

df=pd.DataFrame(l,columns=column_names)
df
```

### *We can now manipulate the data in this dataframe to make inferences*

**Inference 1: Determine how many games on average a Chicago Bulls player played in the 1992-1993 season:**

```python
pd.to_numeric(df['GP']).mean()
```

**Inference 2: Determine how many points per game on average a Chicago Bulls player scored in the 1992-1993 season:**

```python
pd.to_numeric(df['PPG']).mean()
```
